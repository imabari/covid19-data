```
import pdfplumber
import pandas as pd
```

```
import datetime
```

```
def str2date(s: pd.Series) -> pd.Series:

    df = (
        s.str.extract("(\d{1,2})月(\d{1,2})日")
        .rename(columns={0: "month", 1: "day"})
        .fillna(0)
        .astype(int)
    )

    df["year"] = dt_now.year

    tmp = pd.to_datetime(df, errors="coerce")

    df["year"] = df["year"].mask(tmp > dt_now, df["year"] - 1)

    return pd.to_datetime(df, errors="coerce")
```

```
with pdfplumber.open("data.pdf") as pdf:

    dfs = []

    for page in pdf.pages:

        table = page.extract_table()

        df_tmp = pd.DataFrame(table[1:], columns=table[0])

        dfs.append(df_tmp)
```

```
df0 = pd.concat(dfs).reset_index(drop=True)
```

```
# 列名の空白文字削除
df0.columns = df0.columns.str.replace("\s", "")
```

```
# 正規化
for col in df0.select_dtypes(include=object).columns:
    nfkc = df0[col].str.normalize("NFKC")

    if col != "症状、経過、行動歴等":
        nfkc = nfkc.str.replace("\s", "")

    df0[col] = nfkc
```

```
df0
```

# 日付変換

```
dt_now = datetime.datetime.now()
```

```
df1 = df0.copy()
```

```
df2 = (
    df1["症状、経過、行動歴等"]
    .str.extractall("(\d{1,2}月\d{1,2}日)?(.+)\n?")
    .rename(columns={0: "日付", 1: "内容"})
)
```

```
df2["日付"] = str2date(df2["日付"]).fillna(method="ffill")
```

```
df2["内容"] = df2["内容"].str.strip()
```

```
df2["発症"] = df2["日付"].where(df2["内容"].str.contains("発症"))
df2["検体採取"] = df2["日付"].where(df2["内容"].str.contains("検体採取"))
df2["陽性判明"] = df2["日付"].where(df2["内容"].str.contains("陽性判明"))
```

```
df2.to_csv("yousei.csv")
```

```
df2.dropna(thresh=3, inplace=True)
```

```
df2.head(20)
```

```
df3 = (
    df2.droplevel(1)
    .reset_index()
    .groupby("index")
    .agg({"発症": "first", "検体採取": "first", "陽性判明": "first"})
)
```

```
df3
```

```
# 結合
df = df1.join(df3)
```

```
# 発症
df[df["症状、経過、行動歴等"].str.contains("発症") & df["発症"].isna()]
```

```
# 検体採取
df[df["症状、経過、行動歴等"].str.contains("発症") & df["発症"].isna()]
```

```
# 陽性判明
df[df["症状、経過、行動歴等"].str.contains("陽性判明") & df["陽性判明"].isna()]
```
